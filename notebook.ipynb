```python
import pandas as pd
import numpy as np

# Display options
pd.set_option("display.max_columns", 100)
pd.set_option("display.width", 200)
```

```python
files = {
    "results_official": "data/circuit-of-the-americas/COTA/Race 1/00_Results GR Cup Race 1 Official_Anonymized.CSV",
    "results_provisional": "data/circuit-of-the-americas/COTA/Race 1/03_Provisional Results_Race 1_Anonymized.CSV",
    "results_by_class": "data/circuit-of-the-americas/COTA/Race 1/05_Provisional Results by Class_Race 1_Anonymized.CSV",
    "endurance_sections": "data/circuit-of-the-americas/COTA/Race 1/23_AnalysisEnduranceWithSections_Race 1_Anonymized.CSV",
    "weather": "data/circuit-of-the-americas/COTA/Race 1/26_Weather_Race 1_Anonymized.CSV",
    "best10_laps": "data/circuit-of-the-americas/COTA/Race 1/99_Best 10 Laps By Driver_Race 1_Anonymized.CSV",
    "lap_time": "data/circuit-of-the-americas/COTA/Race 1/COTA_lap_time_R1.csv",
    "lap_start": "data/circuit-of-the-americas/COTA/Race 1/COTA_lap_start_time_R1.csv",
    "lap_end": "data/circuit-of-the-americas/COTA/Race 1/COTA_lap_end_time_R1.csv",
}
```

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 200)

def load_csv(path, name):
    print(f"\nLoading: {name}")
    try:
        df = pd.read_csv(path, sep=';')
        print(f"Shape: {df.shape}")
        print(df.head())
        return df
    except Exception as e:
        print("Error:", e)
        return None

results_provisional = load_csv(files["results_provisional"], "results_provisional")
results_by_class = load_csv(files["results_by_class"], "results_by_class")
endurance_sections = load_csv(files["endurance_sections"], "endurance_sections")
weather = load_csv(files["weather"], "weather")
best10 = load_csv(files["best10_laps"], "best10_laps")
lap_time = load_csv(files["lap_time"], "lap_time")
lap_start = load_csv(files["lap_start"], "lap_start")
lap_end = load_csv(files["lap_end"], "lap_end")
results_official = load_csv(files["results_official"], "results_official")
```

```python
def clean_columns(df):
    df.columns = (
        df.columns
        .str.lower()
        .str.replace("*", "", regex=False)
        .str.replace(" ", "_")
        .str.replace("__", "_")
    )
    return df

results_provisional = clean_columns(results_provisional)
results_by_class = clean_columns(results_by_class)
endurance_sections = clean_columns(endurance_sections)
weather = clean_columns(weather)
best10 = clean_columns(best10)
lap_time = clean_columns(lap_time)
lap_end = clean_columns(lap_end)
lap_start = clean_columns(lap_start)
results_official = clean_columns(results_official)

```

```python
for df in [lap_time, lap_end]:
    if "timestamp" in df.columns:
        df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")

    if "meta_time" in df.columns:
        df["meta_time"] = pd.to_datetime(df["meta_time"], errors="coerce")
```

### Feature Engineering

```python
# copy endurance sections for lap data
laps = endurance_sections.copy()

# rename for clarity
laps = laps.rename(columns={
    "driver_number": "driver_id",
    "lap_number": "lap"
})

# convert lap to int
laps["lap"] = pd.to_numeric(laps["lap"], errors="coerce").astype("Int64")
```
